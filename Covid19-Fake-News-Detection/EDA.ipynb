```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data0 = pd.read_csv('/content/drive/MyDrive/Makale-29-4-23/Dataset0.csv')
data1 = pd.read_csv('/content/drive/MyDrive/Makale-29-4-23/Dataset1.csv')
data2 = pd.read_csv('/content/drive/MyDrive/Makale-29-4-23/Dataset2.csv')
data3 = pd.read_csv('/content/drive/MyDrive/Makale-29-4-23/Dataset3.csv')
data4 = pd.read_csv('/content/drive/MyDrive/Makale-29-4-23/Dataset4.csv')
```

```python
def plot_eda_for_dataset(data, dataset_name):
  """Plots EDA for a given dataset."""

  # Basic EDA: Plot distribution of the label_enc column
  plt.figure(figsize=(6, 4))
  sns.countplot(x='label_enc', data=data)
  plt.title(f'Distribution of Label Encodings for {dataset_name}')
  plt.xlabel('Label Encodings')
  plt.ylabel('Count')
  plt.show()

  # Length of processed tweets
  data['tweet_length'] = data['processed_tweet'].apply(len)

  # Plot distribution of tweet lengths
  plt.figure(figsize=(6, 4))
  sns.histplot(data['tweet_length'], bins=30, kde=True)
  plt.title(f'Distribution of Tweet Lengths for {dataset_name}')
  plt.xlabel('Tweet Length')
  plt.ylabel('Frequency')
  plt.show()

  # Plot boxplot of tweet lengths by label
  plt.figure(figsize=(8, 6))
  sns.boxplot(x='label_enc', y='tweet_length', data=data)
  plt.title(f'Boxplot of Tweet Lengths by Label for {dataset_name}')
  plt.xlabel('Label Encodings')
  plt.ylabel('Tweet Length')
  plt.show()

    # 1. Word Count
  data['word_count'] = data['processed_tweet'].apply(lambda x: len(str(x).split()))

  plt.figure(figsize=(6, 4))
  sns.histplot(data['word_count'], bins=30, kde=True)
  plt.title(f'Distribution of Word Counts for {dataset_name}')
  plt.xlabel('Word Count')
  plt.ylabel('Frequency')
  plt.show()

  # 2. Correlation between tweet length and word count
  plt.figure(figsize=(6, 4))
  sns.scatterplot(x='tweet_length', y='word_count', data=data)
  plt.title(f'Correlation between Tweet Length and Word Count for {dataset_name}')
  plt.xlabel('Tweet Length')
  plt.ylabel('Word Count')
  plt.show()

  # 3. Distribution of tweet lengths by label and another parameter (e.g., 'another_column')
  # Replace 'another_column' with an actual column name from your dataset
  if 'another_column' in data.columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='label_enc', y='tweet_length', hue='another_column', data=data)
    plt.title(f'Boxplot of Tweet Lengths by Label and Another Column for {dataset_name}')
    plt.xlabel('Label Encodings')
    plt.ylabel('Tweet Length')
    plt.show()

  # 4. Pairplot for relevant columns (e.g., 'tweet_length', 'word_count', 'label_enc')
  # Replace 'relevant_columns' with your desired columns
  relevant_columns = ['tweet_length', 'word_count', 'label_enc']
  if all(col in data.columns for col in relevant_columns):
    sns.pairplot(data[relevant_columns], hue='label_enc')
    plt.show()
```

```python
# Plot EDA for each dataset in a separate cell
plot_eda_for_dataset(data0, 'Dataset0')
```

```python
plot_eda_for_dataset(data1, 'Dataset1')
```

```python
plot_eda_for_dataset(data2, 'Dataset2')
```

```python
plot_eda_for_dataset(data3, 'Dataset3')
```

```python
plot_eda_for_dataset(data4, 'Dataset4')
```
